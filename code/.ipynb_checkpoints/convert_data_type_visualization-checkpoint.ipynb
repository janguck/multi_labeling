```python
import xlrd
import pickle
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data_dr = '../input/data.xlsx'
```

```python
workbook = xlrd.open_workbook(data_dr)
worksheet = workbook.sheet_by_index(0)
nrows = worksheet.nrows
data_list = [worksheet.row_values(num) for num in range(nrows)]
data_list = np.array(data_list)
```

```python
non_check = data_list[1:,2:].astype(float).sum(axis=1)
```

```python
non_check[non_check>0] = 100
non_check[non_check==0] = 1
non_check[non_check==100] = 0
```

```python
pd_data = pd.DataFrame(data_list[1:,1:],columns=data_list[0][1:])
```

```python
pd_data['아무것도 없어'] = non_check
```

```python
column_index = {}
index = 0
for i in pd_data.columns[1:]:
    column_index[index] = i
    index += 1
```

```python
multi_label = []
cnt = 0
for i in pd_data.values:
    multi_label.append([])
    for j in range(len(i[1:])):
        if i[1:][j] == 1.0 or i[1:][j]=='1.0':
            multi_label[cnt].append(column_index[j])
    multi_label[cnt].sort()
    cnt += 1
multi_label = [" ".join(i) for i in multi_label]
```

```python
pd_data['multilabel'] = multi_label
```

```python
multi_count = pd_data['multilabel'].value_counts().tolist()
```

```python
multi_keys = pd_data['multilabel'].value_counts().keys().tolist()
```

```python
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=path, size=25)
```

```python
import matplotlib as mpl
font_name = mpl.font_manager.FontProperties(fname=path).get_name()
mpl.rc('font', family=font_name)
```

# multilabeling 기준 분포표

```python
for i in range(0,77,7):
    plt.figure(figsize=(25,3))
    plt.bar(multi_keys[i:i+7],multi_count[i:i+7], align='center',color='g',width=0.4)
    plt.xlabel('데이터 이름',fontproperties=fontprop)
    plt.ylabel('개수',fontproperties=fontprop)
    plt.show()
```

# multiclass 기준 분포표

```python
label_name = ['변이', '유전', '경쟁', '자원', '차별적생존', '집단 변화', '목적론', '사용', '의지', '적응','아무것도 없어']
y_data = pd_data[label_name].values
y_data = y_data.astype(float)
label_value = y_data.sum(axis=0)
```

```python
plt.figure(figsize=(25,3))
plt.bar(label_name,label_value, align='center',color='g',width=0.4)
plt.xlabel('데이터 이름',fontproperties=fontprop)
plt.ylabel('개수',fontproperties=fontprop)
plt.show()
```

```python
x_data = pd_data['응답'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.5, random_state=42)
```

```python
train_multi_label = []
cnt = 0
for i in y_train:
    train_multi_label.append([])
    for j in range(len(i)):
        if i[j] == 1.0 or i[j]=='1.0':
            train_multi_label[cnt].append(column_index[j])
    train_multi_label[cnt].sort()
    cnt += 1
train_multi_label = [" ".join(i) for i in train_multi_label]
```

```python
sampling_train = pd.DataFrame(train_multi_label)
sampling_train['train'] = pd.DataFrame(train_multi_label)
```

```python
train_multi_count = sampling_train['train'].value_counts().tolist()
train_multi_keys = sampling_train['train'].value_counts().keys().tolist()
```

```python
for i in multi_keys:
    if i not in train_multi_keys:
        train_multi_keys.append(i)
        train_multi_count.append(0)   
```

```python
dict_train_multi_keys = {}
for i,j in zip(train_multi_keys,train_multi_count):
    dict_train_multi_keys[i]=j
```

```python
test_multi_label = []
cnt = 0
for i in y_test:
    test_multi_label.append([])
    for j in range(len(i)):
        if i[j] == 1.0 or i[j]=='1.0':
            test_multi_label[cnt].append(column_index[j])
    test_multi_label[cnt].sort()
    cnt += 1
test_multi_label = [" ".join(i) for i in test_multi_label]
```

```python
sampling_test = pd.DataFrame(test_multi_label)
sampling_test['test'] = pd.DataFrame(test_multi_label)
```

```python
test_multi_count = sampling_test['test'].value_counts().tolist()
test_multi_keys = sampling_test['test'].value_counts().keys().tolist()
```

```python
for i in multi_keys:
    if i not in test_multi_keys:
        test_multi_keys.append(i)
        test_multi_count.append(0) 
```

```python
dict_test_multi_keys = {}
for i,j in zip(test_multi_keys,test_multi_count):
    dict_test_multi_keys[i]=j
```

```python
refresh_train_multi_keys = []
refresh_test_multi_keys = []
refresh_train_multi_values = []
refresh_test_multi_values = []

for i in multi_keys:
    refresh_train_multi_keys.append(i)
    refresh_test_multi_keys.append(i)
    refresh_train_multi_values.append(dict_train_multi_keys[i])
    refresh_test_multi_values.append(dict_test_multi_keys[i])
```

```python
for i in range(0,77,7):
    plt.figure(figsize=(25,3))
    plt.bar(list(range(7)),refresh_train_multi_values[i:i+7], align='center',color='g',width=0.3,label='train')
    plt.bar(np.array(list(range(7)))+0.4,refresh_test_multi_values[i:i+7], align='center',color='r',width=0.3,label='test')
    plt.xticks(list(range(7)),tuple(refresh_train_multi_keys[i:i+7]))
    plt.xlabel('데이터 이름',fontproperties=fontprop)
    plt.ylabel('개수',fontproperties=fontprop)
    plt.legend()
    plt.show()
```
